# Data Analysis

### Univariate Analysis:

#### Numerical Univariate

1. describe funtion
2. distribution plot
3. boxplot

```python
def numerical_univariate(df, col):
  print("Output of describe function")
  print(df[col].describe())
  print("Output of distribution plot")
  sns.distplot(df[col])
  plt.show()
  print("Output of boxplot")
  sns.boxplot(df[col])
  plt.show()
```

```python
numerical_univariate(df, 'price_per_sqft')
```

Boxplot:
1. lower edge of box : 25th percentile or 1st quartile
2. upper edge of box : 75th percentile or 3rd quartile
3. middle line in box : median or 50th percentile or 2nd quartile
4. IQR (inter quartile range) : height of the box = 75th percentile - 25th percentile
4. lower whisker or lower threshold : 25th percentile - 1.5 * IQR
5. Upper whisker : 75th percentile + 1.5 * IQR
6. Outliers : All data points that lie beyond the whiskers

```python
def remove_outliers(df, col):

  # 25th percentile or 1st quartile
  Q1 = df[col].quantile(0.25)

  # 75th percentile or 3rd quartile
  Q3 = df[col].quantile(0.75)

  # IQR
  IQR = Q3-Q1

  # lower whisker or lower threshold
  lower_threshold = Q1 - 1.5*IQR

  # Upper whisker
  upper_threshold = Q3 + 1.5*IQR

  # remove all values less than lower threshold and greater than upper threshold
  return df[(df[col] > lower_threshold) & (df[col] < upper_threshold)]
```

```python
df_without_outliers = remove_outliers(df, 'price_per_sqft')
```

```python
numerical_univariate(df_without_outliers, 'price_per_sqft')
```

```python
pearson_correlation = df[numerical].corr()
```

Pearson's correlation quantifies the linear correlation between two numerical variables.

- It lies between -1 and 1
- negative correlation indicates that if the value of one variable is increased by 1, the other will decrease by a magnitude equal to the pearson's correaltion coefficient
- positive correlation indicates that if the value of one variable is increased by 1, the other will increase by a magnitude equal to the pearson's correaltion coefficient


```python
pearson_correlation
```

```python
sns.heatmap(pearson_correlation, annot = True)
```
